## CLoud Run Analytics

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

### Container instances Count

```python
df_container = pd.read_csv("cloud_run/Container_instance_count_cr.csv")
```

```python
df_container["Time"] = df_container.Time.apply(lambda x:x[:-len(" GMT-0400 (Eastern Daylight Time)")])
df_container = df_container[df_container.Time>="Sun May 08 2022 13:00:30"].copy()
```

```python
# %matplotlib widget
plt.figure(figsize = (20,10))
ax = sns.lineplot(x="Time", y="value", hue='variable', markers=True, data=pd.melt(df_container, ['Time']))
ax.tick_params(axis='x', rotation=90, labelsize=5);
```
